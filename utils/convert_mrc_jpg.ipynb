```python
import mrcfile
import numpy as np
import cv2
import math
import glob
```

```python
image_path = sorted(list(glob.glob('/bml/Rajan_CryoEM/Processesed_Datasets/Datasets/10028/train/images/*.mrc')))
output_dir = '/bml/ashwin/ViTPicker/data_col/10028/train/'
```

```python
def transform(image):
    i_min = image.min()
    i_max = image.max()

    image = ((image - i_min)/(i_max - i_min)) * 255
    return image.astype(np.uint8)
```

```python
 for single_image in image_path:
    image = mrcfile.read(single_image)
    image = image.T
    image = np.rot90(image)
    image_name = single_image.split("/")[-1].split(".")[0]
   #  print(path)
    image = transform(np.array(image))
    height, width = image.shape
    output_path = output_dir + image_name + '.jpg'
    print(output_path)
    cv2.imwrite(output_path, image)
```
